```python
for i in range(1,51):
    if i % 3 == 0 and i % 5 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
    else:
        print(f"{i}")
```

```python
for i in range(1,51):
    fizz = 'Fizz' if i % 3 == 0 else ''
    buzz = 'Buzz' if i % 5 == 0 else ''
    print(f'{fizz}{buzz}' or i)
```

```python
[print(x) for x in ["FizzBuzz" if y % 15 == 0 else ("Fizz" if y % 3 == 0 else ("Buzz" if y % 5 == 0 else y)) for y in range(1,100)]]
```

```python
print(["FizzBuzz" if y % 15 == 0 else ("Fizz" if y % 3 == 0 else ("Buzz" if y % 5 == 0 else y)) for y in range(1,100)])
```

```python
print(*map(lambda i: 'Fizz'*(not i%3) + 'Buzz'*(not i%5) or i, range(1,51)), sep='\n')
```
