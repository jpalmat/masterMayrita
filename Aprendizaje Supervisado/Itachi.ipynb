```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from skimage import io


img = io.imread('images/itachi.jpg')
new_img = img[0:500, 100:600]
plt.imshow(img, cmap='gray')
plt.show()

# crop, flip vertical y rotacin de -90 grados.
rows, cols, ch = new_img.shape
new_img = cv2.flip(new_img, 0)
M = cv2.getRotationMatrix2D((cols/2,rows/2),angle=-90,scale=1)
new_img = cv2.warpAffine(new_img,M,(cols,rows))
plt.imshow(new_img, cmap='gray')
plt.show()
```

```python
# crop, flip horizontal y rotacin de 90 grados.
new_img = img[0:500, 100:600]
rows, cols, ch = new_img.shape
new_img = cv2.flip(new_img, 1)
M = cv2.getRotationMatrix2D((cols/2,rows/2),angle=90,scale=1)
new_img = cv2.warpAffine(new_img,M,(cols,rows))
plt.imshow(new_img, cmap='gray')
plt.show()
```

```python
# crop, rotacin de -90 grados y flip vertical
new_img = img[0:500, 100:600]
rows, cols, ch = new_img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),angle=-90,scale=1)
new_img = cv2.warpAffine(new_img,M,(cols,rows))
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img, cmap='gray')
plt.show()
```

```python
# crop, rotacin de -90 grados y flip horizontal
new_img = img[0:500, 100:600]
rows, cols, ch = new_img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),angle=-90,scale=1)
new_img = cv2.warpAffine(new_img,M,(cols,rows))
new_img = cv2.flip(new_img, 1)
plt.imshow(new_img, cmap='gray')
plt.show()
```
